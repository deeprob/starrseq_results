```python
def check_sde(ser, meta_exp, libs):
    data_dict = dict()
    for lib in libs:
        data_dict[f"{lib}_abc_sde"] = ""
        data_dict[f"{lib}_nearest_sde"] = ""
    coord = ser.chrom_coord
    if not pd.isnull(ser.abc_gene):
        genes = np.array(ser.abc_gene.split("|"))
        sdes = meta_exp.loc[meta_exp.gene_id.isin(genes), [f"{lib}_padj" for lib in libs]].values
        genes = genes.reshape(len(genes),1).repeat(len(libs), 1)
        sde_genes = np.where(sdes<0.01, genes, "")
        for i,row in enumerate(sde_genes.T):
            data_dict[f"{libs[i]}_abc_sde"] = "|".join(row).strip("|")
    if not pd.isnull(ser.nearest_gene):
        genes = np.array(ser.nearest_gene.split("|"))
        sdes = meta_exp.loc[meta_exp.gene_id.isin(genes), [f"{lib}_padj" for lib in libs]].values
        genes = genes.reshape(len(genes),1).repeat(len(libs), 1)
        sde_genes = np.where(sdes<0.01, genes, "")
        for i,row in enumerate(sde_genes.T):
            data_dict[f"{libs[i]}_nearest_sde"] = "|".join(row).strip("|")
    return pd.Series(data_dict)
```

```python
def create_activity_expression_df(meta_activity_df, meta_expression_df, target_gene_col, libraries, act_column_suff="_act", exp_column_suff="_exp"):
    activity_df = meta_activity_df.loc[~meta_activity_df[target_gene_col].isna()]
    activity_df[target_gene_col] = activity_df[target_gene_col].str.split("|")
    activity_df = activity_df.explode(target_gene_col)
    activity_df = activity_df.merge(meta_expression_df, left_on=target_gene_col, right_on="gene_name", suffixes=(act_column_suff, exp_column_suff))
    return activity_df
```

```python
def parse_gtf_attributes(attr):
    attr_dict = dict()
    for iattr in attr.split(";"):
        if iattr:
            k,v = iattr.strip().split()
            attr_dict[k] = v.strip('"')
    return pd.Series(data=attr_dict)

def parse_gtf_file(gtf_file):
    # read gtf
    gtf_df = pd.read_csv(
        gtf_file, 
        sep="\t", 
        comment="#", 
        header=None,
        names=["chrm", "source", "feature", "start", "end", "score", "strand", "frame", "attribute"]
        )
    # only keep genes
    gtf_df = gtf_df.loc[gtf_df.feature=="gene"]
    # parse the attributes column to get gene id and gene name
    gtf_df = gtf_df.merge(gtf_df.attribute.apply(parse_gtf_attributes), left_index=True, right_index=True)
    gtf_df = gtf_df.loc[:, ["chrm", "start", "end", "gene_id", "gene_name", "strand", "gene_type"]].reset_index(drop=True)
    return gtf_df
```

```python
gtf_df = parse_gtf_file(gtf_file)
```

```python
def make_in_bed_format(gtf_df):
    gtf_df["feature_start"] = gtf_df.start.apply(lambda x: max(x-5000, 0)) 
    gtf_df["feature_end"] = gtf_df.start + 5000
    gtf_df["thickStart"] = gtf_df.start
    gtf_df["thickEnd"] = gtf_df.start
    gtf_df = gtf_df.loc[:, ["chrm", "feature_start", "feature_end", "gene_name", "gene_id", "strand", "thickStart", "thickEnd"]]
    return gtf_df
```
